Vamos a realizar un an√°lisis exploratorio de datos, primero realizamos una ingesta de los datos desde las diversas fuentes, para posteriormente aplicar las transformaciones pertinentes.

```python
#importamos librerias necesarias

import pandas as pd
import numpy as np
import io
import requests
```

```python
#Obtenemos los datasets con la libreria requests y pandas.
#Dataset amazon prime titles

url_amzn = 'https://raw.githubusercontent.com/HX-FAshur/PI01_DATA05/main/Datasets/amazon_prime_titles.csv'
bajada_datos_amzn = requests. get(url_amzn). content
amzn_df = pd. read_csv(io. StringIO(bajada_datos_amzn. decode('utf-8')))

#Dataset disney plus titles

url_disney = 'https://raw.githubusercontent.com/HX-FAshur/PI01_DATA05/main/Datasets/disney_plus_titles.csv'
bajada_datos_disney = requests. get(url_disney). content
disney_df = pd. read_csv(io. StringIO(bajada_datos_disney. decode('utf-8')))


#Dataset hulu titles

url_hulu = 'https://raw.githubusercontent.com/HX-FAshur/PI01_DATA05/main/Datasets/hulu_titles.csv'
bajada_datos_hulu = requests. get(url_hulu). content
hulu_df = pd. read_csv(io. StringIO(bajada_datos_hulu. decode('utf-8')))


#Dataset netflix titles

url_netflix = 'https://raw.githubusercontent.com/HX-FAshur/PI01_DATA05/main/Datasets/netflix_titles.json'
bajada_datos_netflix = requests. get(url_netflix). content
netflix_df = pd. read_json(io. StringIO(bajada_datos_netflix. decode('utf-8')))
```

A√±adimos un nueva columna llamada "plataforma" la cual tendr√° el nombre de su la respectiva plataforma del dataset.

```python
amzn_df['plataforma']='amazon'
disney_df['plataforma']='disney'
hulu_df['plataforma']='hulu'
netflix_df['plataforma']='netflix'
```

En la columna 'show_id' a√±adimos el prefijo de la plataforma para diferenciarlos m√°s adelante.

```python
amzn_df['show_id']='amzn'+ amzn_df['show_id']
disney_df['show_id']='dis'+ disney_df['show_id']
hulu_df['show_id']='hu'+ hulu_df['show_id']
netflix_df['show_id']='net'+ netflix_df['show_id']
```

```python
# Normalizacion de columnas
amzn_df.columns = ['Pelicula_id','Tipo', 'Titulo', 'Director', 'Elenco', 'Pais', 'Fecha_a√±adida', 'A√±o_lanzamiento', 'Clasificacion', 'Duracion', 'Listada_en', 'Descripcion','Plataforma']
disney_df.columns = ['Pelicula_id','Tipo', 'Titulo', 'Director', 'Elenco', 'Pais', 'Fecha_a√±adida', 'A√±o_lanzamiento', 'Clasificacion', 'Duracion', 'Listada_en', 'Descripcion','Plataforma']
hulu_df.columns = ['Pelicula_id','Tipo', 'Titulo', 'Director', 'Elenco', 'Pais', 'Fecha_a√±adida', 'A√±o_lanzamiento', 'Clasificacion', 'Duracion', 'Listada_en', 'Descripcion','Plataforma']
netflix_df.columns = ['Pelicula_id','Tipo', 'Titulo', 'Director', 'Elenco', 'Pais', 'Fecha_a√±adida', 'A√±o_lanzamiento', 'Clasificacion', 'Duracion', 'Listada_en', 'Descripcion','Plataforma']
```

borramos los espacios a los costados de la columna 'Fecha_a√±adida' y le cambiamos el tipo a datetime

```python
amzn_df['Fecha_a√±adida']=amzn_df['Fecha_a√±adida'].str.lstrip()
amzn_df['Fecha_a√±adida']=amzn_df['Fecha_a√±adida'].str.rstrip()
amzn_df['Fecha_a√±adida']=pd.to_datetime(amzn_df['Fecha_a√±adida'], format= '%B %d, %Y')

disney_df['Fecha_a√±adida']=disney_df['Fecha_a√±adida'].str.lstrip()
disney_df['Fecha_a√±adida']=disney_df['Fecha_a√±adida'].str.rstrip()
disney_df['Fecha_a√±adida']=pd.to_datetime(disney_df['Fecha_a√±adida'], format= '%B %d, %Y')

hulu_df['Fecha_a√±adida']=hulu_df['Fecha_a√±adida'].str.lstrip()
hulu_df['Fecha_a√±adida']=hulu_df['Fecha_a√±adida'].str.rstrip()
hulu_df['Fecha_a√±adida']=pd.to_datetime(hulu_df['Fecha_a√±adida'], format= '%B %d, %Y')

netflix_df['Fecha_a√±adida']=netflix_df['Fecha_a√±adida'].str.lstrip()
netflix_df['Fecha_a√±adida']=netflix_df['Fecha_a√±adida'].str.rstrip()
netflix_df['Fecha_a√±adida']=pd.to_datetime(netflix_df['Fecha_a√±adida'], format= '%B %d, %Y')
```

En la columna duraci√≥n tambi√©n se eliminan los espacios laterales

```python
amzn_df['Duracion']=amzn_df['Duracion'].str.lstrip()
amzn_df['Duracion']=amzn_df['Duracion'].str.rstrip()

disney_df['Duracion']=disney_df['Duracion'].str.lstrip()
disney_df['Duracion']=disney_df['Duracion'].str.rstrip()

hulu_df['Duracion']=hulu_df['Duracion'].str.lstrip()
hulu_df['Duracion']=hulu_df['Duracion'].str.rstrip()

netflix_df['Duracion']=netflix_df['Duracion'].str.lstrip()
netflix_df['Duracion']=netflix_df['Duracion'].str.rstrip()
```

Ahora separamos la columna 'duracion' por min

```python
sep_amzn=amzn_df['Duracion'].str.split(" ", n = 1, expand = True) 
sep_hulu=hulu_df['Duracion'].str.split(" ", n = 1, expand = True)
sep_disney=disney_df['Duracion'].str.split(" ", n = 1, expand = True)
sep_netflix=netflix_df['Duracion'].str.split(" ", n = 1, expand = True)
```

```python
amzn_df['duracion_int']=sep_amzn[0]
amzn_df['duracion_tipo']=sep_amzn[1]

disney_df['duracion_int']=sep_disney[0]
disney_df['duracion_tipo']=sep_disney[1]

hulu_df['duracion_int']=sep_hulu[0]
hulu_df['duracion_tipo']=sep_hulu[1]

netflix_df['duracion_int']=sep_netflix[0]
netflix_df['duracion_tipo']=sep_netflix[1]
```

```python
amzn_df
```

Ahora concatenamos los 4 datasets.

```python
concat_1 = pd.concat([amzn_df, disney_df], axis=0)
concat_2=pd.concat([hulu_df,netflix_df], axis=0)
movies_df=pd.concat([concat_1,concat_2], axis=0)
```

```python
movies_df
```

normalizamos la letra del dataset final.

```python
movies_df['Pelicula_id']=movies_df['Pelicula_id'].str.lower()
movies_df['Tipo']=movies_df['Tipo'].str.lower()
movies_df['Titulo']=movies_df['Titulo'].str.lower()
movies_df['Director']=movies_df['Director'].str.lower()
movies_df['Elenco']=movies_df['Elenco'].str.lower()
movies_df['Pais']=movies_df['Pais'].str.lower()
movies_df['Clasificacion']=movies_df['Clasificacion'].str.lower()
movies_df['Duracion']=movies_df['Duracion'].str.lower()
movies_df['Listada_en']=movies_df['Listada_en'].str.lower()
movies_df['Descripcion']=movies_df['Descripcion'].str.lower()
movies_df['Plataforma']=movies_df['Plataforma'].str.lower()
movies_df['duracion_tipo']=movies_df['duracion_tipo'].str.lower()
```

```python
movies_df['duracion_tipo'].replace({'seasons':'season'}, inplace=True)
```
